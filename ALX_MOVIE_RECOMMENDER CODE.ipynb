```python
# Extract year from movie titles

movies['year'] = movies['title'].str.extract(r'\((\d{4})\)', expand=False)
movies['year'] = pd.to_numeric(movies['year'], errors='coerce')

# Handle movies with no year

movies['year'].fillna(0, inplace=True)

# Check the first few rows

print(movies.head())
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse

# Prepare the data for Surprise
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(train[['userId', 'movieId', 'rating']], reader)

# Split the data into training and validation sets
trainset, validationset = train_test_split(data, test_size=0.2)

# Define the SVD model
model = SVD()

# Train the model
model.fit(trainset)

# Evaluate the model
predictions = model.test(validationset)
rmse(predictions)
```

```python
# Function to get predictions for test data
def predict_ratings(model, test_data):
    predictions = []
    for index, row in test_data.iterrows():
        user_id = row['userId']
        movie_id = row['movieId']
        rating = model.predict(user_id, movie_id).est
        predictions.append(rating)
    return predictions

# Get predictions
test['rating'] = predict_ratings(model, test)

# Prepare submission file
test['Id'] = test['userId'].astype(str) + '_' + test['movieId'].astype(str)
submission = test[['Id', 'rating']]
submission.to_csv('submission.csv', index=False)
```
